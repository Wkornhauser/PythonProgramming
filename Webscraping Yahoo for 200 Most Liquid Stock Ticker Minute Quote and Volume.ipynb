### Declare a function to append rows to a prexisting csv

```python
def append_list_as_row(file_name, list_of_elem):
    # Open file in append mode
    with open(file_name, 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow(list_of_elem)
```

### Time to write a big ass loop to go through and scrape each url

```python
def perform_scrape():
    
    date = datetime.datetime.now().strftime('%d-%b-%Y')
    time = datetime.datetime.now().strftime('%H:%M')
    price_volume_list = []
    price_volume_list.append(date)
    price_volume_list.append(time)

    for i in range(len(stocks)):
        # sets the url for scraping 
        url = 'https://ca.finance.yahoo.com/quote/' + str(stocks.iloc[i][0])
    

        try:
            # takes the source code from the url
            source_code = requests.get(url).text
    
            # creates soup object that will be used to parse data
            soup = BeautifulSoup(source_code, 'lxml')
            current_price = soup.find('span', class_="Trsdu(0.3s) Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(b)").text
            print(f'The current price of {stocks.iloc[i][0]} is {current_price}')
            table = soup.find('table', class_='W(100%)')
            days_volume = table.find('span', attrs={'data-reactid': '43'}).text
            print(f'The current volume traded is {days_volume}')
        
            days_volume = days_volume.replace(',','')
            days_volume = int(days_volume)
        
            price_volume_list.append(current_price)
            price_volume_list.append(days_volume)
        
        
        except AttributeError:
            i-=1

    append_list_as_row( 'Trial.txt', price_volume_list)
    return  
```

```python
perform_scrape()
```
