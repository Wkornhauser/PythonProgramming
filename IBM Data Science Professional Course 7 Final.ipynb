```python
df_3 = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_3.drop(['IncidntNum', 'Category', 'Descript', 'DayOfWeek', 'Date', 
           'Time', 'Resolution', 'Address', 'X', 'Y', 'Location', 'PdId'] , axis=1, inplace=True)
df_3['Count']=1
df_3 = df_3.groupby('PdDistrict').count()
df_3 = df_3.reset_index()
df_3
```

```python
import folium

# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42

sf_map = folium.Map(location=[latitude, longitude], zoom_start=12, tiles='Mapbox Bright')

sf_geo = r'san-francisco.geojson' # geojson file

sf_map.choropleth(
    geo_data=sf_geo,
    data=df_3,
    columns=['PdDistrict', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Count'
)



sf_map
```
