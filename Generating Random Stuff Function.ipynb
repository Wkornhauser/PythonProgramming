```python
def generate_random_stuff(seed=None):
    random.seed(seed)
    results = []
    
    for _ in range(5):
        results.append(random.randint(0,5))
        
    characters = list('abc')
    random.shuffle(characters)
    results.append(characters)
    
    for _ in range(5):
        results.append(random.gauss(0,1))
        
    return results
```

```python
generate_random_stuff()
```

```python
def freq_analysis(lst):
    return {k: lst.count(k) for k in set(lst)}
```

```python
lst = [random.randint(0,10) for _ in range(10000000)]
```

```python
print(lst)
```

```python
d = freq_analysis(lst)
```

```python
d
```

```python
total = sum(d.values())
```
